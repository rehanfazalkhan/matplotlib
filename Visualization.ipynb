# Matplotlib Python

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np


x=np.linspace(0,10,20)
y=np.random.randint(0,10,20)
y=np.sort(y)

plt.xlabel('xAxis')
plt.ylabel('yAxis')
plt.title('random plot')
plt.plot(x,y,color='black')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(x,y,'g')
```

```python
fig,ax=plt.subplots(1,2)
ax[0].plot(x,y,'b')
ax[1].plot(x,y,'r')
```

```python
fig,ax=plt.subplots(1,2)
for axes in ax:
    axes.plot(x,y)
```

```python
fig,ax=plt.subplots(1,2)
col=['r','g']

for i,axes in enumerate(ax):
    axes.plot(x,y,col[i])
fig.tight_layout()
```

```python
fig,ax=plt.subplots(figsize=(4,4))
ax.plot(x,y,'r',label='y*y')
ax.plot(x,y*y,'b',label='y')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('random')
ax.legend()
```

```python
fig,ax=plt.subplots()
ax.plot(x,y,'b--',linewidth=2)
```

```python
fig,ax=plt.subplots(1,3,figsize=(8,4))
ax[0].plot(x,y,x,y*y,'r')
ax[1].plot(x,y**2,'b')
ax[1].set_ylim([0,10])
```

```python
plt.scatter(x,y)
```

```python
plt.bar(x,y)
```
